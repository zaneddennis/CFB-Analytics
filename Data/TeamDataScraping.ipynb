```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.sports-reference.com/cfb/years/2018-ratings.html"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
```

```python
rows = soup.findAll("tr")
rows = rows[2:]

for r in rows:
    print(r)
    print()
```
