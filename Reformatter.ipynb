```python
import pandas as pd

df = pd.read_csv("Data/teams_aPPD.tsv", sep="\t", index_col="team")
df = df[["AP", "off_aPPD", "def_aPPD", "net_aPPD"]].copy().round(2)
df
```

```python
for s in ["net", "off", "def"]:
    if s != "def":
        out_df = df.sort_values(by=s+"_aPPD", ascending=False).reset_index()
    else:
        out_df = df.sort_values(by=s+"_aPPD").reset_index()
    out_df.index = out_df.index+1
    out_df.to_csv(s + "_aPPD_asTable.txt", sep="|",  index=True, index_label="rank")
```
