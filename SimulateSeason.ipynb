```python
import pandas as pd

df = pd.read_json("Data/allgames" + str(season) + ".json")
df["fbsonly"] = "no"
for index, row in df.iterrows():
    if row["home_conference"] != None and row["away_conference"] != None:
        df.at[index, "fbsonly"] = "yes"

df = df[df.fbsonly == "yes"]
df = df.set_index("id")
df = df[["start_date", "away_team", "away_points", "home_team", "home_points"]]

df = df.copy()
df
```

```python
for index, row in df.iterrows():
    sd = row["start_date"]
    year = sd.split("-")[0]
    month = sd.split("-")[1]
    day = sd.split("-")[2][:2]
    
    df.at[index, "start_date"] = year+month+day
df
```
