```python
import pandas as pd

df = pd.read_csv("Data/teams_aPPD.tsv", sep="\t", index_col="team")
df = df[["aOPPD", "aDPPD", "net_aPPD"]].copy().sort_values(by="net_aPPD", ascending=False)
df
```

```python
outfile = open("aPPD_asTable.txt", "w")

for index, row in df.iterrows():
    outfile.write(index + " | " + str(row["aOPPD"]) + " | " + str(row["aDPPD"]) + " | " + str(row["net_aPPD"]) + "\n")
```
